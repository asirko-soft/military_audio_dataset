```python
import os
import subprocess

# Set environment variable for CUDA device
os.environ["CUDA_VISIBLE_DEVICES"] = "0"  # Use the first GPU available on Kaggle

# Define model, seed, and tag values
model = "cnn10"
seed = "1"
tag = f"bs64_lr1e-3_ep150seed{seed}"

# Define the command with all parameters
command = [
    "python", "main.py",
    "--tag", tag,
    "--dataset", "military",
    "--seed", seed,
    "--n_cls", "7",
    "--epochs", "150",
    "--batch_size", "64",
    "--optimizer", "adam",
    "--learning_rate", "1e-3",
    "--weight_decay", "1e-6",
    "--cosine",
    "--model", model,
    "--n_mels", "128",
    "--ma_update",
    "--ma_beta", "0.5",
    "--from_sl_official",
    "--audioset_pretrained",
    "--method", "ce",
    "--print_freq", "100",
    "--data_folder", "/kaggle/input/mad-dataset-military-audio-dataset/"
]

# Run the command and display live output
with subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True) as process:
    for line in process.stdout:
        print(line, end="")  # Print each line of stdout as it comes
    for line in process.stderr:
        print(line, end="")  # Print each line of stderr as it comes

# Check for any errors after the process ends
process.wait()
if process.returncode != 0:
    print(f"Process ended with return code {process.returncode}")
```
